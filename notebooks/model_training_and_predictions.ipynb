# Train Test split

```python
train_data ,test_data = tsfm.train_test_split(data ,0.8)
```

# ARIMA model Training

```python
arima_fitted , forecast_arima= am.arima_model(train_data ,test_data)
```

# SARIMA model Training

```python
sarima_fitted , forecast_sarima = sm.sarima_model(train_data,test_data)
```

# Model evaluation

```python
import evalutaion_metrics as em
```

## ARIMA MODEL EVALUATION

```python
mae, rmse, mape = em.evaluate_model(test_data['TSLA'], forecast_arima)
print(f"MAE: {mae}, RMSE: {rmse}, MAPE: {mape}")
```

## ARIMA forecasting and actual data visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,7))
plt.plot(test_data['TSLA'].index, test_data['TSLA'], label='Actual')
plt.plot(test_data['TSLA'].index, forecast_arima, label='Forecast')
plt.title("Forecast vs Actuals")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
